```python
import pandas as pd
import numpy as np
data = pd.read_csv('/content/result_source.csv')
data
```

```python
import requests
from bs4 import BeautifulSoup
import re
url = data['url'][0]
# 記号の削除
code_regex = re.compile('[\t\s!"#$%&\'\\\\()*+,-./:;；：<=>?@[\\]^_`{|}~○｢｣「」〔〕“”〈〉'\
    '『』【】＆＊（）＄＃＠？！｀＋￥¥％♪…◇→←↓↑｡･ω･｡ﾟ´∀｀ΣДｘ⑥◎©︎♡★☆▽※ゞノ〆εσ＞＜┌┘]')
# 数字の削除
num_regex = re.compile('\d+,?\d*')
#日本語チェック用の正規表現
japanese_regex = re.compile('[\uFF01-\uFF0F\uFF1A-\uFF20\uFF3B-\uFF40\uFF5B-\uFF65\u3000-\u303F]+')

def process_url(url):
  response = requests.get(url)
  soup = BeautifulSoup(response.text, "html.parser")
  paragraphs = soup.find_all("p")
  s = str(paragraphs)
  s_list = s.split('p')
  target_texts = []
  for item in s_list:
      item = code_regex.sub("", item)
      item = num_regex.sub("0", item)
      if japanese_regex.search(item):
        target_texts.append(item)
  return target_texts

# 最初の50行のURLに対して処理を適用し、結果を一つのリストにまとめる
all_target_texts = data['url'].iloc[:10].apply(process_url).sum()

#listの要素を結合する
all_target_texts = ' '.join(all_target_texts)
#文単位で再度リストに格納する
all_target_texts = all_target_texts.split('。')
print(all_target_texts)
```

```python
#参照対象URLへ同様の操作を行う
test = process_url("https://zenn.dev/fufufukakaka/articles/068e3adef7fa8e")
test = ' '.join(test)
test = test.split('。')
print(test)
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
embeds = []
for t in all_target_texts:
    embeds.append(embeddings.embed_query(t))
target_embeds = embeds
```

```python
import faiss
import numpy as np

# インデックスの生成
index = faiss.IndexFlatL2(3072)
print(index.is_trained)
print(type(index))

# 対象テキストの追加
index.add(np.array(target_embeds).astype('float32'))
print(index.ntotal)
```

```python
#testに対してfor文で近傍探索を行う
for i in test:
  listed_i = []
  listed_i.append(i)
  in_embeds = embeddings.embed_documents(listed_i)
  D, I = index.search(np.array(in_embeds).astype('float32'), 1)
  # print(D)
  # print(I)
  for j in I[0]:
    if D>1:
      print(i)
      print(all_target_texts[j])
      print(D)
      print("------------------")
```
